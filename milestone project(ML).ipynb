# Calculate Total Attendance

```python
session_cols = [f'session {i}' for i in range(1, 41)]
df_980['Total_Attendance'] = df_980[session_cols].sum(axis=1)
```

```python
df_980[['Student', 'Total_Attendance']].head()
```

# Calculate Attendance Percentage

```python
df_980['Attendance_Percentage'] = (
    df_980['Total_Attendance'] / 40
) * 100
```

```python
df_980[['Student',
        'Total_Attendance',
        'Attendance_Percentage']].head()
```

# Create Certification Column

```python
df_980["Certification"] = df_980["Total_Attendance"].apply(
    lambda x: "Certified" if x >= 32 else "Not Certified"
)
```

```python
df_980[[
    "Student",
    "Total_Attendance",
     "Attendance_Percentage",
    "Certification"
]]
```
